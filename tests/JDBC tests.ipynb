## Second table
Let's now try to read another table while fixing the types using sql.
Actually, they all work now!

```kotlin
import java.sql.DriverManager

val connection = DriverManager.getConnection(url, username, password)
val statement = connection.createStatement()

// Get column names and types
val rs = statement.executeQuery(
    """
    SELECT column_name, data_type
    FROM information_schema.columns
    WHERE table_name = 'film'
    """.trimIndent()
)

// now they all work :D
val unsupportedTypes = setOf<String>(
    //"character varying", // varchar
    //"tsvector", // tsvector
    //"USER-DEFINED", // mpaa_rating == enum
    //"ARRAY" // text[]
)

val selectQuery = buildString {
    append("SELECT ")
    while (rs.next()) {
        val colName = rs.getString("column_name")
        val dataType = rs.getString("data_type")

        // Check if the data type of the column is an unsupportedTypes
        // cast column type to TEXT if so
        if (dataType in unsupportedTypes) {
            append("CAST($colName AS TEXT) AS ");
        }
        append(colName)
        append(",")
    }
    // Remove trailing comma and append table name
    setLength(length - 1)
    append(" FROM film")
}.also(::DISPLAY)

val resultSet = statement.executeQuery(selectQuery)

""
```

## Big table

This will probably fail, but let's read a table with 31 million rows.

Yup, that was too much, let's limit it to 1 mil.

1 million rows now seems possible! maybe Nikita updated the columns part to handle it :)


```kotlin
val df = DataFrame.readSqlTable(connection, "color_dates", limit = 1_000_000)
df
```

```kotlin
import kotlinx.datetime.toKotlinLocalDateTime

@DataSchema
data class RgbColor(val r: Short, val g: Short, val b: Short)

val df2 = df
    .convert { time_in_seconds }.with { it?.toLocalDateTime()?.toKotlinLocalDateTime() } // nullable :/
    .convert { color }.with {
        (it ?: "").removeSurrounding("rgb(", ")") // should not be nullable :/
            .split(", ")
            .map { it.toShort() }
            .let { (r, g, b) -> RgbColor(r, g, b) }
    }
    .unfold { color }

df2
```

Let's try to calculate the average color per date. It should be roughly the same, since they were randomly generated.

```kotlin
val df3 = df2
    .split { time_in_seconds }.by { listOf(it.date, it.time) }.into("date", "time")
    .groupBy("date").into("colors")
```

```kotlin
val df4 = df3.convert { colors }.with { 
    it.meanFor { color.select { cols(r, g, b) } }
}
df4
```
